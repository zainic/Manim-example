```python
%%manim -qm -v WARNING listed

config.media_width = "80%"
import sympy as sp
import numpy as np
import manimpango

class listed(Scene):
    def construct(self):
      X = manimpango.list_fonts()
      for i in range(len(X)):
        lis = Text(X[i], font=X[i])
        self.add(lis)
        self.wait(0.25)
        self.remove(lis)
      self.wait(0.25)
```

```python
%%manim -qm -v WARNING intro

config.media_width = "80%"
import sympy as sp
import numpy as np
import manimpango

class intro(Scene):
    def construct(self):
      def polar_function1(theta):
        r = 3*np.sin(5*theta)
        vector = [r*np.cos(theta), r*np.sin(theta), 0]
        return vector
      def polar_function2(theta):
        r = 1.5*np.sin(5*theta)
        vector = [r*np.cos(theta), r*np.sin(theta), 0]
        return vector
      ax = Axes(
            x_range=[-7, 7, 1],
            y_range=[-4, 4, 1],
            x_length=15,
            y_length=8,
            axis_config={"include_tip": False}
      )
      func1 = VGroup()
      for i in range(10):
          func1.add(ParametricFunction(lambda t : ax.coords_to_point(*polar_function1(t)), 
                                    t_range = np.array([2*TAU/10, 3*TAU/10]), 
                                    fill_opacity=1).set_color(BLUE).set_color_by_gradient(PURE_GREEN, BLUE_E))
      func2 = VGroup()
      for i in range(10):
          func2.add(ParametricFunction(lambda t : ax.coords_to_point(*polar_function2(t)), 
                                    t_range = np.array([2*TAU/10, 3*TAU/10]), 
                                    fill_opacity=1).set_color(BLUE_A).set_color_by_gradient(GREEN_A, BLUE_C))
      circ = Circle(radius=0.5, color=GREEN).set_fill(YELLOW).set_opacity(1)
      self.play(DrawBorderThenFill(func1), 
                DrawBorderThenFill(func2),
                FadeIn(circ)
                ,run_time=2)
      logo = Text("Phoenix", font="TEX Gyre Chorus", weight=BOLD).scale(2)
      COLOR = [BLUE_C, TEAL_C, GREEN_C, RED_A, RED_C, MAROON_C, PURPLE_C]
      for i, letter in enumerate(logo):
            letter.set_color(COLOR[i])
            letter.set_stroke(width=2.5, color=WHITE)
      self.play(Write(logo, lag_ratio=0.1),
                Rotate(func1[1], TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[2], 2*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[3], 3*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[4], 4*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[5], 5*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[6], 6*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[7], 7*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[8], 8*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func1[9], 9*TAU/10, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[0], 0 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[1], TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[2], 2*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[3], 3*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[4], 4*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[5], 5*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[6], 6*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[7], 7*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[8], 8*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                Rotate(func2[9], 9*TAU/10 + TAU/20, OUT, ax.coords_to_point(0,0,0)),
                run_time=4)
      self.wait()
      self.play(ShrinkToCenter(func1, run_time=1),
                ShrinkToCenter(func2, run_time=1),
                ShrinkToCenter(circ, run_time=1.5, lag_ratio=0.5), 
                Unwrite(logo, run_time=3, lag_ratio=0))
```

```python
%%manim -qm -v WARNING scene_1

config.media_width = "80%"
import sympy as sp
import numpy as np
import manimpango

class scene_1(Scene):
    def construct(self):
        equation1 = MathTex(r"ax^2 + bx + c").scale(2)
        equation2 = MathTex(r"ax^2 + bx + c", r"= 0").scale(2)
        # self.add(equation1)
        self.play(Write(equation1), run_time=2)
        self.play(ReplacementTransform(equation1, equation2))
        self.wait(2)
        self.play(equation2.animate.scale(0.75).to_edge(UP))
        quadratic_eq = MathTex(r"x_{1,2} = \dfrac{-b\pm \sqrt{b^2 - 4ac}}{2a}").scale(1.5).shift(0.5*DOWN)
        #self.add(quadratic_eq)
        self.play(Write(quadratic_eq), run_time=3)
        text1 = Text("?").scale(3)
        groups1 = VGroup(equation2, quadratic_eq)
        self.wait()
        self.play(ReplacementTransform(equation2, text1), ReplacementTransform(quadratic_eq, text1))
        self.wait()
```

```python
%%manim -qm -v WARNING bts_0

config.media_width = "80%"
import sympy as sp
import numpy as np
import manimpango

class bts_0(Scene):
    def construct(self):
        def prototype():
            Head = Circle(radius=0.25).set_color(BLUE).set_fill(BLUE_A).set_opacity(1)
            Body = Ellipse(width=0.5, height=1).set_color(BLUE).set_fill(BLUE_A).set_opacity(1)
            L_arm = Ellipse(width=0.25, height=0.7).rotate((180-45) * DEGREES).set_color(BLUE).set_fill(BLUE_A).set_opacity(1)
            R_arm = Ellipse(width=0.25, height=0.7).rotate(45 * DEGREES).set_color(BLUE).set_fill(BLUE_A).set_opacity(1)
            L_leg = Ellipse(width=0.25, height=0.8).rotate((180-10) * DEGREES).set_color(BLUE).set_fill(BLUE_A).set_opacity(1)
            R_leg = Ellipse(width=0.25, height=0.8).rotate(10 * DEGREES).set_color(BLUE).set_fill(BLUE_A).set_opacity(1)
            upper = Head
            arm = VGroup(L_arm, R_arm).arrange(RIGHT, buff=0.4).align_to(Body, UP).shift(DOWN*0.2)
            middle = VGroup(arm, Body)
            bottom = VGroup(L_leg, R_leg).arrange(RIGHT, buff=0)
            person = VGroup(bottom, middle, upper).arrange(UP, buff=-0.1)
            return person
        
        people1 = prototype()
        self.add(people1)
```
