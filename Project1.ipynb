```python
%%manim -qm -v WARNING scene_2

config.media_width = "80%"
import sympy as sp
import numpy as np

class scene_2(Scene):
    def construct(self):
        def SieveOfEratosthenes(n):
            prime = [True for i in range(n+1)]
            p = 2
            while(p * p <= n):
              if (prime[p] == True):
                for i in range(p * p, n + 1, p):
                  prime[i] = False
              p += 1
            X = []
            for p in range(2, n):
              if prime[p]:
                X.append(p)
            return X
        PRIME1 = SieveOfEratosthenes(200000)
        PRIME2 = np.zeros([len(PRIME1)])
        PRIME2[1:len(PRIME1)] = PRIME1[0:len(PRIME1)-1]
        DIFF_PRIME = np.uint(PRIME1 - PRIME2)
        PRIME = np.zeros([np.uint(np.max(DIFF_PRIME))])
        for i in range(1, np.uint(np.max(DIFF_PRIME)+1)):
          for j in DIFF_PRIME:
            if j==i:
              PRIME[i-1] += 1
        PRIME = np.uint(PRIME)
        ax = Axes(
            x_range=[0, np.max(DIFF_PRIME), 1],
            y_range=[0, np.max(PRIME), np.max(PRIME)/6],
            x_length=13,
            y_length=6,
            axis_config={
                "include_tip": False
            }
        )
        self.add(ax)
        for i,rect in enumerate(PRIME):
          poly = [ax.coords_to_point(*[i,0,0]),
                ax.coords_to_point(*[i+1,0,0]),
                ax.coords_to_point(*[i+1,rect,0]),
                ax.coords_to_point(*[i,rect,0])]
          built = Polygon(*poly)
          num = Tex(str(i+1),font_size=9).next_to(built, UP)
          self.add(built, num)
```

```python
%%manim -qm -v WARNING scene_6

config.media_width = "80%"
import sympy as sp
import numpy as np

class scene_6(Scene):
    def construct(self):
        def velocity_x(t, v_vek, ro, m, A, Cd):
            g = 9.80665
            v1,v2 = v_vek[0], v_vek[1]
            vx1 = v2
            vx2 = - (Cd*ro*A*(v2**2))/(2*m)
            return vx1, vx2

        def velocity_y(t, v_vek, ro, m, A, Cd):
            g = 9.80665
            v1,v2 = v_vek[0], v_vek[1]
            vy1 = - v2
            vy2 = g - (Cd*ro*A*(v2**2))/(2*m)
            return vy1, vy2

        def RK4(initial, lower, upper, h, ro, m, A, Cd, diff_func):
            t = np.arange(lower, upper, h)
            n, dim = len(t), len(initial)
            yvek = np.zeros([n, dim])
            k1v, k2v, k3v, k4v = np.zeros([n, dim]), np.zeros([n, dim]), np.zeros([n, dim]), np.zeros([n, dim])
            yvek[0, :] = initial
            for i in range(n-1):
              k1v[i, :] = h * np.array(diff_func(t[i], yvek[i, :], ro, m, A, Cd))
              k2v[i, :] = h * np.array(diff_func(t[i] + (1/2)*h, yvek[i, :] + (1/2)*k1v[i, :], ro, m, A, Cd))
              k3v[i, :] = h * np.array(diff_func(t[i] + (1/2)*h, yvek[i, :] + (1/2)*k2v[i, :], ro, m, A, Cd))
              k4v[i, :] = h * np.array(diff_func(t[i] + h, yvek[i, :] + k3v[i, :], ro, m, A, Cd))
              yvek[i+1, :] = yvek[i, :] + (1/6)*(k1v[i, :] + 2*k2v[i, :] + 2*k3v[i, :] + k4v[i, :])
            return yvek

        def position(x_initial, y_initial, lower, upper, h, ro, m, A, Cd, vx_func, vy_func):
            g = 9.80665
            x = RK4(x_initial, lower, upper, h, ro, m, A, Cd, vx_func)
            y = RK4(y_initial, lower, upper, h, ro, m, A, Cd, vy_func)
            pos = np.zeros([len(y), 3])
            velx = np.zeros([len(x), 3])
            vely = np.zeros([len(y), 3])
            i = 0
            while i < len(y):
              z = 0 
              if y[i,0] >= 0:
                pos[i] = [x[i,0], y[i,0], z]
              elif y[i-1,0]*y[i,0] < 0:
                pos[i] = [x[i,0], 0, z]
                JUM = i+1
              else:
                pos[i] = [x[i,0], 0, z]
              vely[i] = [x[i,0], y[i,1], z]
              i += 1
            i = 0
            while i < len(x):
              z = 0 
              velx[i] = [x[i,0], x[i,1], z]
              i += 1
            Pos = pos[:JUM]
            Vely = vely[:JUM]
            Velx = velx[:JUM]
            return Pos, Velx, Vely

        ax = NumberPlane(
            x_range=[0, 10, 1],
            y_range=[0, 12, 1],
            x_length=6,
            y_length=6,
            background_line_style={
                "stroke_color": BLUE,
                "stroke_width": 3,
                "stroke_opacity": 0.2
            },
            axis_config={
                "include_tip": False,
                "include_numbers": True
            },
            y_axis_config={
                "label_direction":UL
            }
        ).to_edge(LEFT)
        ax2 = NumberPlane(
            x_range=[0, 10, 1],
            y_range=[0, 24, 2],
            x_length=6,
            y_length=6,
            background_line_style={
                "stroke_color": BLUE,
                "stroke_width": 3,
                "stroke_opacity": 0.2
            },
            axis_config={
                "include_tip": False,
                "include_numbers": True
            },
            y_axis_config={
                "label_direction":UL
            }
        ).to_edge(RIGHT)
        lower = 0
        upper = 7
        h = 0.01
        m = 5
        A = PI*(1**2)
        Cd = np.random.uniform(0.5, 0.7)
        ro = 1.81
        g = 9.80665
        y_init = [10, 0]
        x_init = [0, 20]
        POS, VEL_X, VEL_Y = position(x_init, y_init, lower, upper, h, ro, m, A, Cd, velocity_x, velocity_y)
        func = VGroup()
        func.add(ParametricFunction(lambda t : ax.coords_to_point(*POS[np.uint(t)]), 
                                    t_range=np.array([0, len(POS)-2, 1]), use_smoothing=False).set_color(GREEN))
        func.add(ParametricFunction(lambda t : ax2.coords_to_point(*VEL_X[np.uint(t)]), 
                                    t_range=np.array([0, len(VEL_X)-2, 1]), use_smoothing=False).set_color(YELLOW))
        func.add(ParametricFunction(lambda t : ax2.coords_to_point(*VEL_Y[np.uint(t)]), 
                                    t_range=np.array([0, len(VEL_Y)-2, 1]), use_smoothing=False).set_color(BLUE))
        self.add(ax, ax2)
        self.play(Create(func[0], rate_func=rate_functions.linear),
                  Create(func[1], rate_func=rate_functions.linear),
                  Create(func[2], rate_func=rate_functions.linear),
                  run_time=4)
        self.wait()
        sublegend1 = VGroup(
            Line().set_color(GREEN), 
            Tex(": Position", font_size=28)
        ).arrange(RIGHT)
        sublegend2 = VGroup(
            Line().set_color(YELLOW), 
            Tex(": Velocity in x-axis", font_size=28)
        ).arrange(RIGHT)
        sublegend3 = VGroup(
            Line().set_color(BLUE),
            Tex(": Velocity in y-axis", font_size=28)
        ).arrange(RIGHT).align_to(sublegend1, LEFT)
        grup = VGroup(sublegend1, sublegend2, sublegend3).arrange(DOWN)
        legend = SurroundingRectangle(grup).set_color(WHITE)
        G = VGroup(grup, legend).to_corner(UR)
        sublegend1.align_to(sublegend2, LEFT)
        self.play(Create(grup),Create(legend),run_time=2)
        self.wait()
        GRUP_A = VGroup(ax, func[0])
        GRUP_B = VGroup(ax2, func[1], func[2])
        self.play(GRUP_A.animate.scale(0.5), GRUP_B.animate.scale(0.5))
        self.play(GRUP_A.animate.to_corner(UL), GRUP_B.animate.to_corner(DL), run_time=2)
        box_A = SurroundingRectangle(GRUP_A, buff=0.2).set_color(WHITE)
        box_B = SurroundingRectangle(GRUP_B, buff=0.2).set_color(WHITE)
        self.play(FadeIn(box_A), FadeIn(box_B))
        ##################
        self.wait()
        self.play(G.animate.next_to(box_A, RIGHT, buff=0.4).scale(0.75).align_to(box_A, UP))
        eq_a1 = MathTex(r"F_y", r"=", r"W", r"-", r"D_y", font_size=35)
        eq_a2 = MathTex(r"F_y", r"=", r"ma_y", font_size=35)
        eq_a3 = MathTex(r"W", r"=", r"mg", font_size=35)
        eq_a4 = MathTex(r"D_y", r"=", r"\dfrac{\rho {v_y}^2 C_d A_y}{2}", font_size=35)
        eq_a5 = MathTex(r"ma_y", r"=", r"mg", r"-", r"\dfrac{\rho {v_y}^2 C_d A_y}{2}", font_size=35)
        eq_a = VGroup(eq_a1, eq_a2, eq_a3, eq_a4, eq_a5).arrange(DOWN).align_to(G, LEFT)
        eq_a4.align_to(eq_a5, LEFT)
        eq_a3.align_to(eq_a5, LEFT)
        eq_a2.align_to(eq_a5, LEFT)
        eq_a1.align_to(eq_a5, LEFT)
        eq_b1 = MathTex(r"F_x", r"=", r"-", r"D_x", font_size=35)
        eq_b2 = MathTex(r"F_x", r"=", r"ma_x", font_size=35)
        eq_b3 = MathTex(r"D_x", r"=", r"\dfrac{\rho {v_y}^2 C_d A_y}{2}", font_size=35)
        eq_b4 = MathTex(r"ma_x", r"=", r"-", r"\dfrac{\rho {v_x}^2 C_d A_x}{2}", font_size=35)
        eq_b = VGroup(eq_b1, eq_b2, eq_b3, eq_b4).arrange(DOWN).next_to(eq_a, RIGHT, buff=1).align_to(eq_a, UP)
        eq_b4.next_to(eq_a5, RIGHT, buff=1)
        eq_b3.next_to(eq_a4, RIGHT, buff=1)
        eq_b3.align_to(eq_b4, LEFT)
        eq_b2.align_to(eq_b4, LEFT)
        eq_b1.align_to(eq_b4, LEFT)
        eq_a1[1:].align_to(eq_a5[1], LEFT)
        eq_b1[1:].align_to(eq_b4[1], LEFT)
        self.play(Write(eq_a1), Write(eq_b1), run_time=1.5)
        self.wait()
        self.play(ReplacementTransform(eq_a1[0].copy(), eq_a2[0]),
                  ReplacementTransform(eq_b1[0].copy(), eq_b2[0]))
        self.play(Write(eq_a2[1:].align_to(eq_a5[1], LEFT)),
                  Write(eq_b2[1:].align_to(eq_b4[1], LEFT)))
        self.wait()
        self.play(ReplacementTransform(eq_a1[2].copy(), eq_a3[0]))
        self.play(Write(eq_a3[1:].align_to(eq_a5[1], LEFT)))
        self.wait()
        self.play(ReplacementTransform(eq_a1[4].copy(), eq_a4[0]),
                  ReplacementTransform(eq_b1[3].copy(), eq_b3[0]))
        self.play(Write(eq_a4[1:].align_to(eq_a5[1], LEFT)),
                  Write(eq_b3[1:].align_to(eq_b4[1], LEFT)))
        self.wait()
        self.play(ReplacementTransform(eq_a2[2].copy(), eq_a5[0]),
                  ReplacementTransform(eq_b2[2].copy(), eq_b4[0]))
        self.play(Write(eq_a5[1]), Write(eq_b4[1]))
        self.play(ReplacementTransform(eq_a3[2].copy(), eq_a5[2]))
        self.play(Write(eq_a5[3]), Write(eq_b4[2]))
        self.play(ReplacementTransform(eq_a4[2].copy(), eq_a5[4]),
                  ReplacementTransform(eq_b3[2].copy(), eq_b4[3]))
        self.wait()
        self.play(FadeOut(eq_a[:4]), FadeOut(eq_b[:3]), 
                  eq_a5.animate.move_to(eq_a1).align_to(eq_a1, LEFT),
                  eq_b4.animate.move_to(eq_b1).align_to(eq_b1, LEFT))
        self.wait()
        EQ_A = MathTex(r"a_y", r"=", r"g", r"-", r"\dfrac{\rho C_d A_y}{2m}",r"{v_y}^2", font_size=35)
        EQ_A.move_to(eq_a1).align_to(eq_a1, LEFT)
        EQ_B = MathTex(r"a_x", r"=", r"-", r"\dfrac{\rho C_d A_x}{2m}",r"{v_x}^2", font_size=35)
        EQ_B.move_to(eq_b1).align_to(eq_b1, LEFT)
        self.play(ReplacementTransform(eq_a5, EQ_A), ReplacementTransform(eq_b4, EQ_B))
        self.wait()
        EQ_A1 = MathTex(r"\dfrac{d^2s_y}{dt^2}", r"=", r"g", r"-", r"\dfrac{\rho C_d A_y}{2m}",r"\left(\dfrac{ds_y}{dt}\right)^2", font_size=35)
        EQ_A1.next_to(EQ_A, DOWN).align_to(eq_a1, LEFT)
        EQ_B1 = MathTex(r"\dfrac{d^2s_y}{dt^2}", r"=", r"-", r"\dfrac{\rho C_d A_x}{2m}",r"\left(\dfrac{ds_x}{dt}\right)^2", font_size=35)
        EQ_B1.next_to(EQ_B, DOWN).align_to(eq_b1, LEFT)
        X1, Y1 = EQ_A[0].copy(), EQ_B[0].copy()
        X2, Y2 = EQ_A[-1].copy(), EQ_B[-1].copy()
        self.play(X1.animate.move_to(EQ_A1[0]),
                  Y1.animate.move_to(EQ_B1[0]))
        self.play(ReplacementTransform(X1, EQ_A1[0]),
                  ReplacementTransform(Y1, EQ_B1[0]))
        self.play(X2.animate.move_to(EQ_A1[-1]),
                  Y2.animate.move_to(EQ_B1[-1]))
        self.play(ReplacementTransform(X2, EQ_A1[-1]),
                  ReplacementTransform(Y2, EQ_B1[-1]))
        self.play(Write(EQ_A1[1:5]), Write(EQ_B1[1:4]))
        self.wait()
        note1 = Tex("Using RK4 for solving the differential equation", font_size=31).next_to(EQ_A1, DOWN).align_to(EQ_A1, LEFT)
        self.play(Write(note1))
        note2 = MathTex(r"\text{We get } (s_x, s_y) \text{ as position of the object}", font_size=31).next_to(note1, DOWN).align_to(EQ_A1, LEFT)
        self.play(Write(note2.set_color(GREEN)))
        M = SurroundingRectangle(GRUP_A).set_color(GREEN)
        self.wait()
        self.play(Create(M))
        self.play(FadeOut(M))
        self.wait()
        self.play(Unwrite(note1), Unwrite(note2), Unwrite(EQ_A), Unwrite(EQ_B))
        self.play(G.animate.scale(1.25).align_to(box_A, UP).to_edge(RIGHT),
                  EQ_B1.animate.scale(0.8).to_edge(RIGHT),
                  EQ_A1.animate.scale(0.8).next_to(EQ_B1, UP).to_edge(RIGHT))
        self.wait()
        #################
        # h = 0.01
        # m = 5
        # A = PI*(1**2)
        # Cd = np.random.uniform(0.5, 0.7)
        # ro = 1.81
        # g = 9.80665
        # y_init = [10, 0]
        # x_init = [0, 20]
        tex0 = MathTex(r"&\text{Object = Ball}\\",
                       r"&m = 5 \text{ kg}\\",
                       r"&A_x = A_y = \pi \text{ m}^2 \:\:(r=1)\\",
                       r"&Cd = "+str(Cd)+r"\\",
                       r"&g = 9.80665\text{ m/s}^{-1}\\",
                       r"&\rho = 1.81\text{ kg/m}^3 \:(\,air\:density\,)\\",
                       r"&(s_x(0),v_x(0))=(0,20)\\",
                       r"&(s_y(0),v_y(0))=(10,0)\\",
                       font_size=26).to_corner(DR)
        self.play(Write(tex0))
        self.wait()
        self.play(FadeOut(box_A), FadeOut(box_B))
        self.play(FadeOut(GRUP_B), GRUP_A.animate.scale(2).to_corner(DL), run_time=2)
        self.wait()
        dot1 = Dot().set_color(GREEN_A).move_to(ax.coords_to_point(0,10,0))
        text_num1 = MathTex(r"t = ", font_size=30).set_color(GREEN_A)
        text_num2 = MathTex(r"s_x = ", font_size=30).set_color(GREEN_A)
        text_num3 = MathTex(r"s_y = ", font_size=30).set_color(GREEN_A)
        text_num = VGroup(text_num1, text_num2, text_num3).arrange(DOWN).next_to(GRUP_A, RIGHT, buff=1).align_to(tex0, UP)
        text_num2.align_to(text_num3, LEFT)
        text_num1.align_to(text_num3, LEFT)
        num1 = DecimalNumber(0, num_decimal_places=2, font_size=30).next_to(text_num1, RIGHT).set_color(GREEN_A)
        num2 = DecimalNumber(0, num_decimal_places=2, font_size=30).next_to(text_num2, RIGHT).set_color(GREEN_A)
        num3 = DecimalNumber(10, num_decimal_places=2, font_size=30).next_to(text_num3, RIGHT).set_color(GREEN_A)
        Track = ValueTracker(0)
        self.play(FadeIn(dot1), FadeIn(text_num), FadeIn(num1), FadeIn(num2), FadeIn(num3))
        dot1.add_updater(lambda t : t.move_to(ax.coords_to_point(POS[np.uint(Track.get_value())][0], 
                                                                 POS[np.uint(Track.get_value())][1], 
                                                                 0)))
        num1.add_updater(lambda x : x.become(DecimalNumber(np.uint(Track.get_value())*h, 
                                                           num_decimal_places=2, 
                                                           font_size=30
                                                           ).next_to(text_num1, RIGHT).set_color(GREEN_A)))
        num2.add_updater(lambda x : x.become(DecimalNumber(POS[np.uint(Track.get_value())][0], 
                                                           num_decimal_places=2, 
                                                           font_size=30
                                                           ).next_to(text_num2, RIGHT).set_color(GREEN_A)))
        num3.add_updater(lambda x : x.become(DecimalNumber(POS[np.uint(Track.get_value())][1], 
                                                           num_decimal_places=2, 
                                                           font_size=30
                                                           ).next_to(text_num3, RIGHT).set_color(GREEN_A)))
        self.play(Track.animate(rate_func=rate_functions.linear).set_value(len(POS)-1), run_time=3)
        self.play(FadeOut(dot1), FadeOut(text_num), FadeOut(num1), FadeOut(num2), FadeOut(num3))
        self.play(FadeIn(GRUP_B), GRUP_A.animate.scale(0.5).to_corner(UL), run_time=2)
        self.wait()
        self.play(FadeOut(GRUP_A), GRUP_B.animate.scale(2).to_corner(DL), run_time=2)
        self.wait()
        dot2 = Dot().set_color(YELLOW_A).move_to(ax2.coords_to_point(0,20,0))
        dot3 = Dot().set_color(BLUE_A).move_to(ax2.coords_to_point(0,0,0))
        text_num1 = MathTex(r"t = ", font_size=30).set_color(RED)
        text_num2 = MathTex(r"v_x = ", font_size=30).set_color(YELLOW_A)
        text_num3 = MathTex(r"v_y = ", font_size=30).set_color(BLUE_A)
        text_num = VGroup(text_num1, text_num2, text_num3).arrange(DOWN).next_to(GRUP_B, RIGHT, buff=1).align_to(tex0, UP)
        text_num2.align_to(text_num3, LEFT)
        text_num1.align_to(text_num3, LEFT)
        num1 = DecimalNumber(0, num_decimal_places=2, font_size=30).next_to(text_num1, RIGHT).set_color(RED)
        num2 = DecimalNumber(20, num_decimal_places=2, font_size=30).next_to(text_num2, RIGHT).set_color(YELLOW_A)
        num3 = DecimalNumber(0, num_decimal_places=2, font_size=30).next_to(text_num3, RIGHT).set_color(BLUE_A)
        Track = ValueTracker(0)
        self.play(FadeIn(dot2), FadeIn(dot3), FadeIn(text_num), FadeIn(num1), FadeIn(num2), FadeIn(num3))
        dot2.add_updater(lambda t : t.move_to(ax2.coords_to_point(VEL_X[np.uint(Track.get_value())][0], 
                                                                 VEL_X[np.uint(Track.get_value())][1], 
                                                                 0)))
        dot3.add_updater(lambda t : t.move_to(ax2.coords_to_point(VEL_Y[np.uint(Track.get_value())][0], 
                                                                 VEL_Y[np.uint(Track.get_value())][1], 
                                                                 0)))
        num1.add_updater(lambda x : x.become(DecimalNumber(np.uint(Track.get_value())*h, 
                                                           num_decimal_places=2, 
                                                           font_size=30
                                                           ).next_to(text_num1, RIGHT).set_color(RED)))
        num2.add_updater(lambda x : x.become(DecimalNumber(VEL_X[np.uint(Track.get_value())][1], 
                                                           num_decimal_places=2, 
                                                           font_size=30
                                                           ).next_to(text_num2, RIGHT).set_color(YELLOW_A)))
        num3.add_updater(lambda x : x.become(DecimalNumber(VEL_Y[np.uint(Track.get_value())][1], 
                                                           num_decimal_places=2, 
                                                           font_size=30
                                                           ).next_to(text_num3, RIGHT).set_color(BLUE_A)))
        self.play(Track.animate(rate_func=rate_functions.linear).set_value(len(VEL_X)-1), run_time=3)
        self.play(FadeOut(dot2), FadeOut(dot3), FadeOut(text_num), FadeOut(num1), FadeOut(num2), FadeOut(num3))
        self.play(FadeIn(GRUP_A), GRUP_B.animate.scale(0.5).to_corner(DL), run_time=2)
        self.wait()
```
